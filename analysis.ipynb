```python
import numpy as np
from collections import OrderedDict
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.lines as mlines
from matplotlib.ticker import FormatStrFormatter
```

```python
folder='outputs'
arch=['A100','H200','B200','Bnext']
Mem={'A100': 80, 'H200': 141, 'B200': 192, 'Bnext': 384}
tp_strategy=['1d']#,'2d']
models=['gpt3','gpt3_1T']

def get_stats_best_model(folder,model,arch,tp,loc=0):
    data=np.load(folder+'/exec_'+tp+'_model_'+model+'_config_'+arch+'.npz',allow_pickle=True)
    ngpus=data['ngpus'][()]
    nvs=sorted(list(data['ngpus'][()].keys()))
    confs=OrderedDict()
    throughputs=OrderedDict()
    times=OrderedDict()
    comms=OrderedDict()
    mems=OrderedDict()
    for nv in nvs:
        confs[nv]=[]
        throughputs[nv]=[]
        times[nv]=[]
        comms[nv]=[]
        mems[nv]=[]
        for n in sorted(ngpus[nv]):
            loct = min(loc,len(data['confs'][()][nv][n])-1)
#             print(loct,len(data['confs'][()][nv][n]))
            confs[nv].append(data['confs'][()][nv][n][loct])
            throughputs[nv].append(data['throughputs'][()][nv][n][loct])
            times[nv].append([data['stats'][()][nv][n][loct]['t_com']/data['stats'][()][nv][n][loct]['t'],
                   data['stats'][()][nv][n][loct]['t_comp']/data['stats'][()][nv][n][loct]['t'],
                   data['stats'][()][nv][n][loct]['t_mem']/data['stats'][()][nv][n][loct]['t']])
            comms[nv].append(data['stats'][()][nv][n][loct]['comm_fwd'])
            mems[nv].append(data['stats'][()][nv][n][loct]['mem'])
    
    return nvs,ngpus,confs,throughputs,times,comms,mems


def get_stats_bms(folder,model,arch,tp,lr,er,dr,loc=0):
    ngps=OrderedDict()
    nvss=OrderedDict()
    confs=OrderedDict()
    throughputs=OrderedDict()
    times=OrderedDict()
    comms=OrderedDict()
    mems=OrderedDict()
    for l in lr:
        for e in er:
            for d in dr:
                filename=folder+'/exec_'+tp+'_model_'+model+'_config_'+arch\
                 +'_lfactor_'+str(l)+'_efactor_'+str(e)+'_dfactor_'+str(d)+'.npz'
                print(filename)
                data=np.load(filename,allow_pickle=True)
                ngps[(l,e,d)]=data['ngpus'][()]
                nvss[(l,e,d)]=sorted(list(data['ngpus'][()].keys()))
                
                for nv in nvss[(l,e,d)]:
                    confs[(nv,l,e,d)]=[]
                    throughputs[(nv,l,e,d)]=[]
                    times[(nv,l,e,d)]=[]
                    comms[(nv,l,e,d)]=[]
                    mems[(nv,l,e,d)]=[]
                    for n in sorted(ngps[(l,e,d)][nv]):
                        loct = min(loc,len(data['confs'][()][nv][n])-1)
                        confs[(nv,l,e,d)].append(data['confs'][()][nv][n][loct])
                        throughputs[(nv,l,e,d)].append(data['throughputs'][()][nv][n][loct])
#                         print(nv,n,loct,data['throughputs'][()][nv][n][loct])
                        times[(nv,l,e,d)].append([data['stats'][()][nv][n][loct]['t_com']/data['stats'][()][nv][n][loct]['t'],
                            data['stats'][()][nv][n][loct]['t_comp']/data['stats'][()][nv][n][loct]['t'],
                            data['stats'][()][nv][n][loct]['t_mem']/data['stats'][()][nv][n][loct]['t']])
                        comms[(nv,l,e,d)].append(data['stats'][()][nv][n][loct]['comm_fwd'])
                        mems[(nv,l,e,d)].append(data['stats'][()][nv][n][loct]['mem'])
    
    return nvss,ngps,confs,throughputs,times,comms,mems
```

```python
nvs1d={}
ngpus1d={}
confs1d={}
throughputs1d={}
times1d={}
comms1d={}
mems1d={}
lr=[1,2,3,4]
er=[1]
dr=[1]


for model in models:
    for tp in ['1d']:#tp_strategy:
        for gpu in arch:
            key=(model,tp,gpu)
            nvs1d[key],ngpus1d[key],confs1d[key],throughputs1d[key],\
            times1d[key],comms1d[key],mems1d[key]=get_stats_bms(folder,model,gpu,tp,lr,er,dr,loc=0)
            #get_stats_best_model(folder,model,gpu,tp,0)
            
            
# nvs2d={}
# ngpus2d={}
# confs2d={}
# throughputs2d={}
# times2d={}
# comms2d={}
# mems2d={}

# for model in models:
#     for tp in ['2d']:#tp_strategy:
#         for gpu in arch:
#             key=(model,tp,gpu)
#             nvs2d[key],ngpus2d[key],confs2d[key],throughputs2d[key],\
#             times2d[key],comms2d[key],mems2d[key]=get_stats_best_model(folder,model,gpu,tp,0)
```
